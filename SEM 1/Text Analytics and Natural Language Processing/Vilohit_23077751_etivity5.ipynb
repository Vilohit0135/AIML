```python
positive_words = positive_df["Positive Words"].values
negative_words = negative_df["Negative Words"].values

def analyze_sentiment(text):
    words = text.split()
    positive_count = sum(word in positive_words for word in words)
    negative_count = sum(word in negative_words for word in words)

    if((positive_count + negative_count) == 0):
      sentiment = 'neutral 😐'
      confidence =0
    else:
      confidence = (max(positive_count,  negative_count) / len(words))*100
      if positive_count > negative_count:
          sentiment = 'positive 😊'
      elif negative_count > positive_count:
          sentiment = 'negative 😣'

    print(f'Sentiment {sentiment}')
    print(f'Confidance {confidence}')
```

```python
analyze_sentiment("i am 2-faced ")
```

```python
analyze_sentiment("i am accurate ")
```
